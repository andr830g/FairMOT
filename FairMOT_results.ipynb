```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
logResultsFairMOT = pd.read_csv("logResultsFairMOT.csv")
trackResultsFairMOT = pd.read_csv("trackResultsFairMOT.csv")
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].plot(list(range(1,30+1)), logResultsFairMOT["loss"], color='blue', label="total loss")
ax[0].set_title("Total Loss")
ax[0].set_xlabel("Epochs")
ax[0].set_ylabel("Loss")
ax[0].legend()

det_loss = (np.array(logResultsFairMOT["hm_loss"]) + np.array(logResultsFairMOT["wh_loss"]) + np.array(logResultsFairMOT["off_loss"])).tolist()
ax[1].plot(list(range(1,30+1)), logResultsFairMOT["id_loss"], color='red', label="identity loss")
ax[1].plot(list(range(1,30+1)), det_loss, color="grey", label="detection loss")
ax[1].set_title("Branch Loss")
ax[1].set_xlabel("Epochs")
ax[1].set_ylabel("Loss")
ax[1].legend()

ax[2].plot(list(range(1,30+1)), logResultsFairMOT["hm_loss"], color='orange', label="heatmap loss")
ax[2].plot(list(range(1,30+1)), logResultsFairMOT["wh_loss"], color='green', label="size loss")
ax[2].plot(list(range(1,30+1)), logResultsFairMOT["off_loss"], color='purple', label="offset loss")
ax[2].set_title("Heads Loss")
ax[2].set_xlabel("Epochs")
ax[2].set_ylabel("Loss")
ax[2].legend()

plt.suptitle("FairMOT MOT17half Training Loss", fontsize=18, fontweight="bold")

plt.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].plot(list(range(1,30+1)), trackResultsFairMOT["IDF1"], color='blue')
ax[0].set_title("IDF1")
ax[0].set_xlabel("Epochs")
ax[0].set_ylabel("IDF1")

ax[1].plot(list(range(1,30+1)), trackResultsFairMOT["Ids"], color='red')
ax[1].set_title("IDs")
ax[1].set_xlabel("Epochs")
ax[1].set_ylabel("IDs")

ax[2].plot(list(range(1,30+1)), trackResultsFairMOT["MOTA"], color='green')
ax[2].set_title("MOTA")
ax[2].set_xlabel("Epochs")
ax[2].set_ylabel("MOTA")

plt.suptitle("FairMOT MOT17 Validation Results", fontsize=18, fontweight="bold")

plt.tight_layout()

plt.show()
```
